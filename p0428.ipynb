```python
ebola = pd.read_csv('pandasdata/country_timeseries.csv')
print(ebola.fillna(method='ffill').iloc[0:10, 0:5])

# 결측치가 나타나기 전의 값으로 변경한다. 예를 들면 6행 결측치 가 있을때 
# 5행의 값을 사용하여 결측치를 처리한다. 0, 1행의 경우에는 전의 값이 없기 때문에
# 결측치가 그대로 남아있게 된다. 
```

```python
ebola = pd.read_csv('pandasdata/country_timeseries.csv')
print(ebola.fillna(method='bfill').iloc[0:10, 0:5])

# bfill 로 지정하면 결측치가 나타난 이후의 첫번째 값으로 앞쪽의 결측치가 모두 변경된다. 즉 ffill의 반대 방향으로 처리한다. 하지만 이 방법도 마지막 값이 누락값인 경우에는 처리하지 못한다는 단점이 있다. 
```

```python
print(ebola.interpolate().iloc[0:10, 0:5])

# 위에서의 두가지 방법 ffill  bfill 는 데이터 프레임의 값을 그대로 사용하여 결측치를 처리했으나 interpolate매서드는 누락값 양쪽에 있는 값을 이용하여 중간값을 구해서 결측치를 처리한다. 이렇게 하면 데이터 프레임이 일정한 간격을 유지하고 있는 것처럼 수정할수있다. 
```

```python
# 깔끔한 데이터

# 데이터프레임의 열이 파이썬의 변수와 비슷한 역할을 하게 된다.  예를 들면 ebola데이터프레임의 열은 날짜와 국가들의 데이터를 저장하고 있다. 데이터 프레임은 열이 옆으로 길게 늘어선 형태가 된다. 이것을 '넓은 데이터' 라고 부른다. 

# melt 메서드 - 지정한 열의 데이터를 모두 행으로 정리해 준다. 따라서 데이터를 바라보는 관점을 쉽게 바꿀수가 있다. 

pew = pd.read_csv('pandasdata/pew.csv')
```

```python
pew
# 미국의 소득과 종교에 관한 데이터 셋으로  총 11개의 열이 있다. 
```

```python
# 소득 정보를 행으로 보면 어떨까...

pew_long = pd.melt(pew, id_vars='religion')
pew_long

# id_vars 인자로 지정한 열을 제외한 나머지 모든 소득 열들이 variable 열로 정리되고 소득 정보열의 행데이터도 value 열로 정리되었다. 'religion열을 고정하여 피벗했다' 라고 표현한다. 

```

```python
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')

pew_long
```

```python
# 하나의 열이 여러 의미를 갖고 있는 경우

ebola = pd.read_csv('pandasdata/country_timeseries.csv')
ebola.shape
```

```python
ebola.columns
```

```python
ebola.iloc[:5, [0,1,2,3,10,11]]
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date','Day'])
ebola_long
```

```python
# Cases_Guinea 와같이 2개 이상의 의미를 가지고 있는 열의 이름은 _을 기준으로 
# cases   Guinea 와 같이 분리할 수 있다. 

variable_split = ebola_long.variable.str.split('_')
variable_split
```

```python
type(variable_split)
```

```python
type(variable_split[0])
```

```python
# 0번쨰 인덱스는 cases는 발병 deaths는 죽음을 의미하고
# 1번째 인덱스는 국가명들을 의미한다. 이것들을 분리하여 데이터프레임에 새로운 열로 넣자


status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
```

```python
country_values
```

```python
ebola_long
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long
```
