```python
import pandas as pd

df1 = pd.read_csv('pandasdata/concat_1.csv')
df2 = pd.read_csv('pandasdata/concat_2.csv')
df3 = pd.read_csv('pandasdata/concat_3.csv')
```

```python
# concat() - 위에서 아래 방향으로 연결

row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
# 데이터 프레임에 시리즈 연결

new_row = pd.Series(['n1','n2','n3','n4'])
```

```python
print(pd.concat([df1, new_row]))

# 시리즈가 새로운 행으로 추가 될것을 기대했지만 새로운 열로 삽입이 되면서
# 구조를 오히려 복잡하게 만들고 있다. NaN(누락값) 많이 보인다. 

# 행이 1개라도 반드시 데이터프레임에 담아서 연결하자. 
# 시리즈에는 열이름이 없다. 따라서 새로운 열로 간주하여 0이라는 이름의 열로 
# 추가하게 된것이다. 


```

```python
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])

print(new_row_df)
```

```python
# 방법 1, 데이터 프레임으로 만들어서 데이터프레임끼리  concat으로 연결

pd.concat([df1, new_row_df])
```

```python
# 방법 2, 데이터 프레임으로 만들어서 데이터프레임끼리  append 로 연결
df1.append(new_row_df)
```

```python
# 방법 3, 

data_dict = {'A': 'n1','B':'n2','C':'n3','D':'n4'}

print(df1.append(data_dict, ignore_index=True))

# 마지막 행의 인덱스가 0 이 아닌 4로 변경 ignore_index=True
```

```python
# 새로 삽입된 데이터의 기존 인덱스를 무시하고 삽입된 새로운 인덱스로 대체된다. 

pd.concat([df1, new_row_df], ignore_index=True)
```

```python
# 열방향으로 데이터 연결하자

col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

```python
col_concat['A']
```

```python
col_concat['F'] = ['n1','n2','n3','n4']
col_concat
```

```python
# 열 이름의 일부가 서로 다른 데이터 프레임을 연결하면 어떻게 될까...
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']
```

```python
row_concat2 = pd.concat([df1,df2,df3])
row_concat2

# 새롭게 열이름을 부여하고 3개의 데이터 프레임을 concat으로 연결하였다. 
# 열이름이 정렬되어 연결되었고 데이터프레임에 없는 열이름의 데이터는 전부 누락값으로
# 처리된것을 볼수 있다. 
```

```python
# 데이터프레임의 공통열만 골라서 연결하면 누락값이 발생하지 않는다. 
# 현재 df1, df2,df3 모두의 공통열은 없다. 
```

```python
pd.concat([df1,df3])
```

```python
pd.concat([df1,df3], join='inner', ignore_index=True)
```

```python
pd.concat([df1,df3], axis=1, join='inner')

# 공통행인 0과 2만 출력된다. 


```

```python
# 판다스는 데이터 연결 전용 매서드인 merge를 제공한다. 

person = pd.read_csv('pandasdata/survey_person.csv')
site = pd.read_csv('pandasdata/survey_site.csv')
survey = pd.read_csv('pandasdata/survey_survey.csv')
visited = pd.read_csv('pandasdata/survey_visited.csv')
```

```python
visited_subset = visited.loc[[0,2,6],]
visited_subset
```

```python
site
```

```python
merge1 = site.merge(visited_subset, left_on='name', right_on='site')
merge1
```

```python
merge2 = site.merge(visited, left_on='name', right_on='site')
merge2
```

```python
merge3 = person.merge(survey, left_on='ident', right_on='person')
merge3
```

```python
# 누락값 - 결측치

# 대부분의 데이터에는 누락값이 존재한다.  NaN ( NAN, nan )

from numpy import NaN, NAN, nan
```

```python
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
print(NaN == NaN)  # 값 자체가 없는 존재이기때문에 자신과 비교해도 False
print(NaN == nan)
print(NaN == NAN)
```

```python
# 확인하는 매서드 isnull 을 사용
print(pd.isnull(NaN))
print(pd.isnull(NAN))
print(pd.isnull(nan))
```

```python

print(pd.notnull(NaN))
print(pd.notnull(42))
print(pd.notnull('austin'))
```
